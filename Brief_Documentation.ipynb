## Краткая Документация Бота для Телеграмм-Канала "Я-Инженер"

### Введение

Этот бот предназначен для автоматического комментирования постов в телеграмм-канале "Я-Инженер" с помощью API OpenAI. Он позволяет администраторам канала автоматизировать взаимодействие с пользователями, предоставляя краткие и полезные комментарии.

### Функционал

- **Автоматическое комментирование**: Бот может комментировать посты в канале, используя контекст поста и предыдущие комментарии.
- **Настройка количества ответов**: Администратор может установить максимальное количество комментариев бота на каждый пост с помощью команды `/setlimit`.
- **Включение/выключение ответов**: Бота можно временно отключить или включить с помощью команд `/mute` и `/unmute`.
- **Логирование данных**: Все комментарии и взаимодействия записываются в файл для дальнейшего анализа.

### Команды

| Команда | Описание |
|---------|----------|
| `/start` | Выводит приветственное сообщение и информацию о боте. |
| `/help` | Отображает инструкции по использованию бота. |
| `/setlimit <число>` | Устанавливает максимальное количество комментариев бота на пост. |
| `/mute` | Отключает ответы бота. |
| `/unmute` | Включает ответы бота. |

### Настройка

1. **Добавление бота в канал**: Добавьте бота в качестве администратора в телеграмм-канал и в чат обсуждения.
2. **Использование команд**: Используйте команды для настройки поведения бота.

### Технические детали

- **API OpenAI**: Используется для генерации комментариев.
- **Технологии**: Python, Telegram Bot API, OpenAI API.
- **Логирование**: Данные записываются в CSV-файл для анализа.

### Примеры использования

- **Автоматическое комментирование**: После публикации поста бот автоматически добавляет комментарий.
- **Настройка количества ответов**: Используйте `/setlimit 5`, чтобы ограничить бота до 5 комментариев на пост.
